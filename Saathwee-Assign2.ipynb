```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime as dt
from scipy import stats
from sklearn.preprocessing import LabelEncoder
import statsmodels.api as sm

# 1. Creating the Dataset using Pandas
data = {
    'ID': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Gender': ['F', 'M', 'M', 'M', 'F'],
    'Age': [20, 22, 23, 21, 20],
    'Dept': ['CSE', 'IT', 'CSE', 'ECE', 'IT'],
    'GPA': [3.8, 3.2, 3.5, 3.7, 3.9]
}
df = pd.DataFrame(data)

# 2. Pandas - Understanding the Dataset
print(df.head())       # View the first few rows
print(df.info())       # Overview of data types and non-null values
print(df.describe())   # Summary statistics

# 3. NumPy - Statistical Analysis
mean_age = np.mean(df['Age'])
median_gpa = np.median(df['GPA'])
print(f"Mean Age: {mean_age}, Median GPA: {median_gpa}")

# 4. Matplotlib - Basic Visualization
plt.hist(df['Age'])
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

plt.scatter(df['Age'], df['GPA'])
plt.title('Age vs. GPA')
plt.xlabel('Age')
plt.ylabel('GPA')
plt.show()

# 5. Seaborn - Enhanced Visualization
sns.countplot(x='Gender', data=df)
plt.title('Gender Distribution')
plt.show()

sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# 6. Datetime - Date Manipulation Example
current_year = dt.now().year
df['BirthYear'] = current_year - df['Age']
print(df[['Name', 'BirthYear']])

# 7. Scipy - Finding Mode
mode_gender = stats.mode(df['Gender'])
print(f"Most Common Gender: {mode_gender.mode[0]}")

# 8. Sklearn - Encoding Categorical Data
label_encoder = LabelEncoder()
df['DeptEncoded'] = label_encoder.fit_transform(df['Dept'])
print(df[['Dept', 'DeptEncoded']])

# 9. Statsmodels - Simple Regression
X = df['Age']
Y = df['GPA']
X = sm.add_constant(X)  # Adds a constant term to the predictor
model = sm.OLS(Y, X).fit()
print(model.summary())

#5 VISUALISATION

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# Create the dataset
data = {
    'ID': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Gender': ['F', 'M', 'M', 'M', 'F'],
    'Age': [20, 22, 23, 21, 20],
    'Dept': ['CSE', 'IT', 'CSE', 'ECE', 'IT'],
    'GPA': [3.8, 3.2, 3.5, 3.7, 3.9]
}
df = pd.DataFrame(data)

# 1. Age Distribution (Histogram)
plt.figure(figsize=(6,4))
plt.hist(df['Age'], bins=5, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# 2. GPA Distribution by Department (Boxplot)
plt.figure(figsize=(6,4))
sns.boxplot(x='Dept', y='GPA', data=df, palette='Set3')
plt.title('GPA Distribution by Department')
plt.xlabel('Department')
plt.ylabel('GPA')
plt.grid(True)
plt.show()

# 3. Gender Distribution (Count Plot)
plt.figure(figsize=(6,4))
sns.countplot(x='Gender', data=df, palette='coolwarm')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.grid(True)
plt.show()

# 4. Age vs GPA (Scatter Plot)
plt.figure(figsize=(6,4))
plt.scatter(df['Age'], df['GPA'], color='purple')
plt.title('Age vs. GPA')
plt.xlabel('Age')
plt.ylabel('GPA')
plt.grid(True)
plt.show()


#INSIGHTS
print('''Histogram of Age Distribution: This visualizes the distribution of ages among the students.
Boxplot of GPA by Department: Shows how GPA varies across different departments.
Count Plot of Gender Distribution: Displays the distribution of students by gender.
Scatter Plot of Age vs. GPA: Illustrates the relationship between age and GPA.
Bar Chart of GPA by Department: An Altair-based interactive bar chart showing GPA by department.''')
```
